```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the ETT dataset (adjust the path as needed)
ett_data = pd.read_csv('all_six_datasets//ETT-small//ETTm1.csv')


# Plotting the dependent variables
dependent_variables = ['HUFL', 'HULL', 'MUFL', 'MULL', 'LUFL', 'LULL']

# Plot each dependent variable
for variable in dependent_variables:
    plt.figure(figsize=(12, 6))  # Create a new figure for the plot
    plt.plot(ett_data['date'], ett_data[variable], label=variable)  # Plot the variable
    plt.xlabel('Date')  # Set the x-axis label
    plt.ylabel('Value')  # Set the y-axis label
    plt.title(f'{variable} Over Time')  # Set the plot title
    plt.legend()  # Show the legend
    plt.grid(True)  # Show grid lines
    plt.show()  # Display the plot

# Plot the target variable (OT) separately
plt.figure(figsize=(12, 6))  # Create a new figure for the plot
plt.plot(ett_data['date'], ett_data['OT'], label='OT', color='red')  # Plot the target variable
plt.xlabel('Date')  # Set the x-axis label
plt.ylabel('Value')  # Set the y-axis label
plt.title('OT (Target Variable) Over Time')  # Set the plot title
plt.legend()  # Show the legend
plt.grid(True)  # Show grid lines
plt.show()  # Display the plot
```
